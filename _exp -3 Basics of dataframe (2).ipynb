# Creating Data Frame

```python
import pandas as pd
```

```python
df=pd.DataFrame([[1,2,3,4],[6,7,8,9],[11,12,13,14],[16,17,18,19]],
columns= ["Bia","DBMS","OE","DSS"])
```

# Adding Record(Row)

```python
df2=pd.DataFrame([[21,22,23,24]],
columns= ["Bia","DBMS","OE","DSS"])
```

```python
df3 =pd.concat([df,df2], ignore_index=True)
```

```python
df3
```

# Adding Attribute(coloumn)in a data frame

```python
df3["DM"]=[21,22,23,24,25]
```

```python
df3
```

# Deleting Record From Data Frame

```python
df4=df3.drop(1)
```

```python
df4
```

# Deleting Attribute(Columns) From df3 dataframe

```python
df5=df3.drop(columns=["DM"])
```

```python
df5
```
